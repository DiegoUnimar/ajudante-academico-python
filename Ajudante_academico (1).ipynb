```python
import time
import json
import os

ARQUIVO = "alunos.json"

def carregar_todos():
    if os.path.exists(ARQUIVO):
        with open(ARQUIVO, "r", encoding="utf-8") as f:
            return json.load(f)
    return {}

def salvar_todos(todos_alunos):
    with open(ARQUIVO, "w", encoding="utf-8") as f:
        json.dump(todos_alunos, f, ensure_ascii=False, indent=4)

def digitar(texto, velocidade=0.02):
    for letra in texto:
        print(letra, end="", flush=True)
        time.sleep(velocidade)
    print()

def linha():
    print("\n" + "-" * 40 + "\n")

def mostrar_planner(planner):
    if not planner:
        digitar("Você ainda não criou um planner.", 0.03)
        return
    print("\n" + "-" * 45)
    print(f"{'Dia':<12} | {'Aulas':<25}")
    print("-" * 45)
    for dia, aulas in planner.items():
        print(f"{dia:<12} | {aulas:<25}")
    print("-" * 45 + "\n")


# Início
digitar("---- Olá, estudante! ----", 0.03)
todos_alunos = carregar_todos()

linha()
digitar("Qual seu nome?", 0.03)
usuario = input("\n> ").strip().lower().replace(" ", "_")

if usuario not in todos_alunos:
    todos_alunos[usuario] = {"aluno": {}, "planner": {}, "provas": []}
    digitar(f"\nPerfil criado para '{usuario}'.", 0.03)
else:
    digitar(f"\nBem-vindo(a) de volta, {usuario}!", 0.03)

dados   = todos_alunos[usuario]
planner = dados["planner"]
provas  = dados["provas"]


# Menu
while True:
    linha()
    print("1 - Matrícula")
    print("2 - Organizar planner de aulas")
    print("3 - Inserir data de prova")
    print("4 - Ver planner")
    print("5 - Ver provas")
    print("6 - Sair")
    opcao = input("\n> ")

    if opcao == "1":
        linha()
        nome  = input("Nome completo: ")
        data_nascimento = input("Data de nascimento (dd/mm/aaaa): ")
        idade = 2026 - int(data_nascimento[-4:])
        cidade = input("Cidade: ")
        curso  = input("Curso: ")
        dados["aluno"] = {"nome": nome, "idade": idade, "cidade": cidade, "curso": curso}
        salvar_todos(todos_alunos)
        linha()
        digitar(f"Matrícula salva, {nome}!", 0.03)

    elif opcao == "2":
        linha()
        for dia in ["Segunda", "Terca", "Quarta", "Quinta", "Sexta"]:
            planner[dia] = input(f"Aulas de {dia}: ")
        salvar_todos(todos_alunos)
        digitar("Planner salvo!", 0.03)

    elif opcao == "3":
        linha()
        materia = input("Matéria: ")
        data    = input("Data (dd/mm): ")
        provas.append({"materia": materia, "data": data})
        salvar_todos(todos_alunos)
        digitar("Prova salva!", 0.03)

    elif opcao == "4":
        linha()
        mostrar_planner(planner)

    elif opcao == "5":
        linha()
        if not provas:
            digitar("Nenhuma prova salva.", 0.03)
        else:
            for p in provas:
                print(f"  - {p['materia']} → {p['data']}")

    elif opcao == "6":
        digitar("Até logo!", 0.03)
        break

    else:
        digitar("Opção inválida.", 0.03)
```
